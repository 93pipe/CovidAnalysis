```python
#Step 1: Import all packages needed for running the code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
#Step 2: Read the csv file with the data and create the dataframe. Also, define the path where the file will be saved
url = "./public_dataset_original.csv"
df = pd.read_csv(url, header=0)
path = "./public_dataset_2.csv"
df.to_csv(path)
```

```python
#Step 3: First boxplot that shows the people's level of exposure to COVID by country
X = df['country']
Y = df['exposure']
plt.cla()
sns.boxplot(x=Y, y=X, data=df, orient="h", palette="Pastel1")
plt.title("Boxplot exposure region")
plt.show()
```

```python
#Step 4: Second boxplot that shows people's level of awarness about COVID mortality by country. 
#100% --> people believe COVID is extremely mortality; 0% --> people believe COVID is lowly or nothing mortal
X = df['country']
Y = df['Covid Mortality']
plt.cla()
sns.boxplot(x=Y, y=X, data=df, orient="h", palette="Pastel1")
plt.title("Mortality aware by country")
plt.xlabel("Mortality awarness (%)")
plt.show()
```

```python
#Step 5: Third boxplot that shows people's level of exposure to COVID by industry
X = df['Industry']
Y = df['exposure']
plt.cla()
sns.boxplot(x=Y, y=X, data=df, orient="h", palette="Pastel1")
plt.title("Covid Exposure by Industry")
plt.xlabel("Exposure covid (%)")
plt.show()
```

```python
#Step 6: Fourth boxplot that shows people's beliefs of being infected by industry. 
X = df['Industry']
Y = df['Are_you_infected']
plt.cla()
sns.boxplot(x=Y, y=X, data=df, orient="h", palette="Pastel1")
plt.title("Belief of being infected by indsutry")
plt.xlabel("Belief of being infected (%)")
plt.show()
# %%
X = df['country']
Y = df['Are_you_infected']
plt.cla()
sns.boxplot(x=Y, y=X, data=df, orient="h",palette="Pastel1")
plt.title("Belief of being infected by country")
plt.xlabel("Belief of being infected (%)")
plt.show()

```

```python
#Step 7: Fifth boxplot shows the people's level of awarness. All variable are in terms of percentage.
# Are you infected?
# Infected_yourarea -> How many people you think is infected in your area?
# Infected_no_Symp -> How many people do you think dont develop syntoms?
# Infected_no_hosp -> How many people do you think will develop the illness without requiring hospitalize
# Infected_hosp -> How many people do you think will develop the illness requiring hospitalize
# Covid_Mortality -> How many people you think will die because of covid?

# As we can see, a higher percentage of people thinks that covid will not produce any symptoms rather than COVID will kill someone
Y = df[['Are_you_infected', "Infected_yourarea", "Infected_no_Symp", 'Infected_no_hosp', 'Infected_hosp',
        "Covid Mortality"]]
plt.cla()
sns.boxplot(data=Y, orient="h", palette="Pastel1", labels=[["hola", "s", "s", "s", "s", "s"]])
plt.xlim(0, 100)
plt.title("People awarness")
plt.xlabel("Percentage of awarness (%)")
plt.show()
```
